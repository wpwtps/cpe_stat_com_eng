```python
df2 = df
```

```python
def label_EastAsia (row):
    if row['Region'] == 'East Asia':
        return '1'
    return '0'

def label_SouthAmerica(row):
    if row['Region'] == 'South America':
        return '1'
    return '0'

def label_NorthernEurope(row):
    if row['Region'] == 'Northern Europe':
        return '1'
    return '0'
```

```python
df2.apply (lambda row: label_EastAsia(row), axis=1)
#axis1 == do at a row
```

```python
df2['EastAsia'] = df2.apply (lambda row: label_EastAsia(row), axis=1)
df2['SouthAmerica'] = df2.apply (lambda row: label_SouthAmerica(row), axis=1)
df2['NorthernEurope'] = df2.apply (lambda row: label_NorthernEurope(row), axis=1)
df2
```
