# JOIN

```python
#join region with gas_price

df5 = regions.join(df4[['2012 [YR2012]', '2016 [YR2016]']])
df5.rename(columns={'2012 [YR2012]':'gasPrice2012','2016 [YR2016]':'gasPrice2016'}, inplace=True)
df5.head()
```

```python
# joins region with airtran

df6 = df5.join(df3[['2012 [YR2012]', '2016 [YR2016]']])
df6.rename(columns={'2012 [YR2012]':'airTrans2012','2016 [YR2016]':'airTrans2016'}, inplace=True)
df6.head()
```
