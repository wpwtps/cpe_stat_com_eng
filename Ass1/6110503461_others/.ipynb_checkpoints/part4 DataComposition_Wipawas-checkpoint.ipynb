```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as pyo
import plotly.graph_objs as go
%matplotlib inline
```

```python
df = pd.read_csv('cair2.csv',sep=',', header='infer')
dff2 = df.rename(columns={"Region": "Year"})
dff2
```

```python
df2 =dff2.set_index('Year')
meandf2 = df2.groupby(by="Year").mean()
meandf2
```

```python
dff2.to_csv('df2air.csv',index=False)
```

```python
df_trans = meandf2.T
```

```python
df_trans
```

```python
df_trans.set_index("Year", inplace = True)
df_trans
```

```python
df_trans
```

```python
sns.lineplot(data=df_trans)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
chart = df_trans.plot.line()
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
trace = go.Scatter(x=df.index,
                    y=df_trans,
                    mode='lines',
                    name='time_to_series')
data = [trace]
```

```python
layout = go.Layout(title='airtransport',
                   xaxis = {'title':'Year'},
                   yaxis = {'title':'Country',
                           'range': [560,5800000]}, 
                   hovermode='closest',
                   )
```

```python
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
