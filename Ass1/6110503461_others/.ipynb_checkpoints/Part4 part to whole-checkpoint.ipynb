```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as pyo
import plotly.graph_objs as go
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python

df = pd.read_excel('gdp_per_capita.xlsx')
df
```

```python
regions = pd.read_excel('countries_regions.xlsx')
regions.head()
```

```python
df2 = df
df2.drop([217,218,219,220,221], inplace=True)
```

```python
df2.replace({'..':np.nan}, inplace=True)
df2
```

```python
sns.heatmap(df2.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df3 = regions.join(df2[['2012 [YR2012]', '2016 [YR2016]']])
```

```python
df3
```

```python
df3.rename(columns={'2012 [YR2012]':'gdp2012','2016 [YR2016]':'gdp2016'}, inplace=True)
df3.head()
```

```python
df4 = pd.read_excel('airtransport.xlsx')
df4
```

```python
df5 = df4
df5.drop([217,218,219,220,221], inplace=True)
```

```python
df5.replace({'..':np.nan}, inplace=True)
df5
```

```python
sns.heatmap(df5.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df6 = df3.join(df5[['2012 [YR2012]', '2016 [YR2016]']])
```

```python
df6
```

```python
df6.rename(columns={'2012 [YR2012]':'air2012','2016 [YR2016]':'air2016'}, inplace=True)
df6.head()
```

```python
df7 = df6.dropna()
```

```python
sns.heatmap(df7.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(15,15))
the_grid = gridspec.GridSpec(2, 2)

plt.subplot(the_grid[0, 0],  title='gdp_per_Capital 2012')
sns.barplot(y='Region',x='gdp2012', data=df7, palette='Spectral')
plt.tight_layout(15)

plt.subplot(the_grid[0, 1],  title='air transportation 2012')
sns.barplot(y='Region',x='air2012', data=df7, palette='Spectral')
```

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(15,15))
the_grid = gridspec.GridSpec(2, 2)

plt.subplot(the_grid[0, 0],  title='gdp_per_Capital 2016')
sns.barplot(y='Region',x='gdp2016', data=df7, palette='Spectral')
plt.tight_layout(15)

plt.subplot(the_grid[0, 1],  title='air transportation 2016')
sns.barplot(y='Region',x='air2016', data=df7, palette='Spectral')
```
