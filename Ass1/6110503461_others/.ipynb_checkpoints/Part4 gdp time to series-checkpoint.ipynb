```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as pyo
import plotly.graph_objs as go
%matplotlib inline
```

```python
gdp = pd.read_excel('gdp_per_capita.xlsx',sheet_name='Data')
gdp
```

```python
gdp.drop([217, 218, 219, 220, 221], inplace=True)
gdp
```

```python
gdp2 = gdp.drop(['Series Name','Series Code','Country Code','Country Code'], axis=1)
gdp2
```

```python
gdp2.replace({'..':np.nan}, inplace=True)
gdp2.head()
```

```python
sns.heatmap(gdp2.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
gdp2.drop(['2019 [YR2019]'],axis=1, inplace=True)
gdp2.head()
```

```python
regions = pd.read_excel('countries_regions.xlsx')
regions.head()
```

```python
regions.drop(['ThirdWorld'],axis=1, inplace=True)
regions.head()
```

```python
reggdp = regions.set_index('CountryName').join(gdp2.set_index('Country Name'))

reggdp
```

```python
cgdp = reggdp.dropna()
cgdp
```

```python
cgdp.groupby(by="Region").mean()
```

```python
cgdp.to_csv('cgdp2.csv',index=False)
```

```python
df = pd.read_csv('cgdp2.csv',sep=',', header='infer')
dff2 = df.rename(columns={"Region": "Year"})
dff2
```

```python
df2 =dff2.set_index('Year')
meandf2 = df2.groupby(by="Year").mean()
meandf2
```

```python
df_trans = meandf2.T
```

```python
df_trans
```

```python
sns.lineplot(data=df_trans)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
